```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
import math
```

讀入檔案

```python
img = cv2.imread('./001.jpg',  cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('./002.jpg', cv2.IMREAD_GRAYSCALE)
img.shape
img2.shape
```

進行直方圖平均化 (左為原圖，右為平均化結果，使用numpy hstack合併）

```python
equ = cv2.equalizeHist(img)
res = np.hstack((img,equ))
plt.title('Histogram equalization')
plt.imshow(res, cmap = 'gray')
```

執行X和Y方向的Sobel算子

```python
x = cv2.Sobel(img,cv2.CV_16S,1,0)
y = cv2.Sobel(img,cv2.CV_16S,0,1)
absX = cv2.convertScaleAbs(x)
absY = cv2.convertScaleAbs(y)
dst = cv2.addWeighted(absX,0.5,absY,0.5,0)
plt.title('Sobel operator (x-axis)')
plt.imshow(absX)
```

```python
plt.title('Sobel operator (y-axis)')
plt.imshow(absY)
```

```python
plt.title('Sobel operator')
plt.imshow(dst)
```

最後使用Canny算子處理

```python
gaussian_img = cv2.GaussianBlur(img,(3,3),0)
plt.imshow(gaussian_img)
plt.title('Gaussian')
plt.show()
canny = cv2.Canny(gaussian_img, 50, 150)  

plt.imshow(canny)
plt.title('Canny operator')
plt.show()
```

第二張圖也一樣

```python
%matplotlib inline
equ = cv2.equalizeHist(img2)
plt.imshow(equ, cmap = 'gray')
plt.title('Histogram equalization')
plt.show()
x = cv2.Sobel(img2,cv2.CV_16S,1,0)  
y = cv2.Sobel(img2,cv2.CV_16S,0,1)  
  
absX = cv2.convertScaleAbs(x) 
absY = cv2.convertScaleAbs(y)  
  
dst = cv2.addWeighted(absX,0.5,absY,0.5,0)  
  
plt.imshow(absX)
plt.title('Sobel operator (x-axis)')
plt.show()
plt.imshow(absY)
plt.title('Sobel operator (y-axis)')
plt.show()
plt.imshow(dst)
plt.title('Sobel operator')
plt.show()

gaussian_img = cv2.GaussianBlur(img2,(3,3),0)
plt.imshow(gaussian_img)
plt.title('Gaussian')
plt.show()
canny = cv2.Canny(gaussian_img, 50, 150)  

plt.imshow(canny)
plt.title('Canny operator')
plt.show()
```
